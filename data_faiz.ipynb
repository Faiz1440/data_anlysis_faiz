### remove stopwatch from our description 

```python
pip install nltk
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords = stopwords.words('english')
def remove_stopwords(text):
    words = [word for word in text.split() if word.lower() not in stopwords]
    new_text = " ".join(words)
    return new_text
```

```python
data["listed_in"] = data["listed_in"].apply(lambda x:remove_stopwords(x))
```

```python
data["description"] = data["description"].apply(lambda x:remove_stopwords(x))
```

```python
data["cast"] = data["cast"].apply(lambda x:remove_stopwords(x))
```
